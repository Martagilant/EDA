```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Assuming df is your DataFrame with Amazon_year and Categories columns

# Group by year and category to count occurrences
cat_counts_per_year = df.groupby(['Amazon_year', 'Categories']).size().reset_index(name='Count')

# Get the most popular category for each year
most_popular_per_year = cat_counts_per_year.loc[
    cat_counts_per_year.groupby('Amazon_year')['Count'].idxmax()
]

# Sort by year to ensure chronological display
most_popular_per_year = most_popular_per_year.sort_values('Amazon_year')

# Get unique categories
unique_categories = most_popular_per_year['Categories'].unique()

# Create a color map using a more distinct colormap
# Tab20 provides up to 20 distinct colors that are easily distinguishable
if len(unique_categories) <= 20:
    color_map = plt.cm.tab20
elif len(unique_categories) <= 10:
    color_map = plt.cm.tab10  # Even more distinct with fewer categories
else:
    # For many categories, use a high-contrast colormap
    color_map = plt.cm.hsv

colors = color_map(np.linspace(0, 0.9, len(unique_categories)))  # Avoid full circle in hsv

# Create a dictionary to map each category to a color
category_to_color = {category: colors[i] for i, category in enumerate(unique_categories)}

# Assign colors to the bars based on the category
bar_colors = most_popular_per_year['Categories'].map(category_to_color)

# Plot the results
plt.figure(figsize=(14, 8))
bars = plt.bar(most_popular_per_year['Amazon_year'].astype(str), most_popular_per_year['Count'],
        color=bar_colors, edgecolor='black')

# No text inside bars, only use the legend for category identification


# Create a legend mapping categories to colors
legend_patches = [plt.Rectangle((0, 0), 1, 1, color=category_to_color[cat]) for cat in unique_categories]
plt.legend(legend_patches, unique_categories, loc='upper right', bbox_to_anchor=(1.15, 1))

# Customize plot
plt.title('Most Popular Category Each Year', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.grid(False)

# Save figure
plt.savefig("barplot_bestcat_year.png", dpi=300, bbox_inches='tight')
plt.show()

```

```python
def pinta_distribucion_categoricas(df, columnas_categoricas, relativa=False, mostrar_valores=False):
    num_columnas = len(columnas_categoricas)
    num_filas = (num_columnas // 2) + (num_columnas % 2)

    fig, axes = plt.subplots(num_filas, 2, figsize=(15, 5 * num_filas))
    axes = axes.flatten() 

    for i, col in enumerate(columnas_categoricas):
        ax = axes[i]
        if relativa:
            total = df[col].value_counts().sum()
            serie = df[col].value_counts().apply(lambda x: x / total)
            sns.barplot(x=serie.index, y=serie, ax=ax, palette='viridis', hue = serie.index, legend = False)
            ax.set_ylabel('Frecuencia Relativa')
        else:
            serie = df[col].value_counts()
            sns.barplot(x=serie.index, y=serie, ax=ax, palette='viridis', hue = serie.index, legend = False)
            ax.set_ylabel('Frecuencia')

        ax.set_title(f'Distribución de {col}')
        ax.set_xlabel('')
        ax.tick_params(axis='x', rotation=45)

        if mostrar_valores:
            for p in ax.patches:
                height = p.get_height()
                ax.annotate(f'{height:.2f}', (p.get_x() + p.get_width() / 2., height), 
                            ha='center', va='center', xytext=(0, 9), textcoords='offset points')

    for j in range(i + 1, num_filas * 2):
        axes[j].axis('off')

    plt.grid(False)

    plt.tight_layout()
    plt.savefig("barplot_genre", dpi=300, bbox_inches='tight')
    plt.show()

pinta_distribucion_categoricas(df,["Genres"], relativa= True, mostrar_valores= False)
```
